```python
# Método MLCG(16807, 2^31 -1)

def mlcg(x, n):
  xn = [x]
  for i in range(n):
    xn.append((16807*xn[-1]) % (2**31 -1))
  r = np.array(xn)/(2**31-1)
  return r
```

```python
r = mlcg(10, 10)
r
```

```python
# RAND LCG(69069, 1, 2^32)
def RAND(x, n):
  xn = [x]
  for i in range(n):
    xn.append((69069*xn[-1] + 1) % (2**32))
  r = np.array(xn)/(2**32)
  return r
```

#### Metodo de la Funcion Inversa



$$f(x) = 3x^2 \qquad 0<x<1$$

$$F(X) = x^3 \qquad 0<x<1$$

$$ u = x^3$$

$$ u^{\frac{1}{3}} = (x^3)^{\frac{1}{3}}$$

$$ F^{-1}(u) =  u^{\frac{1}{3}}$$

```python
u = np.random.rand(1000)
x = u**(1/3)
plt.hist(x, density=True, rwidth=0.9)
x1 = np.linspace(0,1,100)
f1 = 3*x1**2
plt.plot(x1,f1,'--')
```

**Distribución exponencial**

$$f(x) = \lambda e^{-\lambda x}$$

$$F(x) = 1 - e^{-\lambda x}$$

$$ u = 1 -\lambda  e^{-\lambda x}$$

$$ 1 - u = \lambda  e^{-\lambda x}$$

$$ \ln (1 - u) =   -\lambda x$$

$$ -\frac{1}{\lambda}\ln (1 - u) =  x$$

$$ F^{-1}(u) = -\frac{1}{\lambda}\ln (1 - u)$$

```python
lamb = 3
u = np.random.rand(1000)
x = -(1/lamb)*np.log(1-u)
plt.hist(x, density=True, rwidth=0.9)
x1 = np.linspace(0,2,100)
f1 = lamb*np.exp(-lamb*x1)
plt.plot(x1,f1,'--')
```

### Generación de una Distribución Beta

Se desea generar la siguiente variable aleatorio

$$X\sim Beta(\alpha=2, \beta=2)$$

La función de densidad está dada por:

$$f(x)=6x(1-x) \quad 0\leq x \leq 1$$

Para ello se utilizará una distribución uniforme para $g(x)$

$$g(x) = 1 \quad 0\leq x \leq 1$$


```python
def f(x):
  x = np.array(x)
  return 6*x*(1-x)
```

```python
def g(x):
  x = np.array(x)
  return np.ones_like(x)
```

```python
c = 1.6
x = np.linspace(0,1,100)
fx = f(x)
gx = g(x)
plt.plot(x,fx,'-b')
plt.plot(x,c*gx,'-r')
```

Una variable aleatoria X obtenida de $g(x)$ es aceptada si:

$$\frac{f(x)}{cg(x)} = \frac{6x(1-x)}{1.6} = 3.75x(1-x)>u$$

```python
# Paso 1.
y = np.random.rand()  # Se genera número aleatorio con $g$
print("número generado: ", y)
criterio = 3.75*y*(1-y)
print("f/(cg) = ", criterio)
u = np.random.rand() # se genera número aleatorio para u
print("uniforme = ", u)
if criterio>u:
  print('Se acepta:', y)
else:
  print('Se rechaza:',y)
```

```python
def aceptacion_rechazo(N = 10):
  k = 0
  j = 0
  X = []
  while k<N: 
    j += 1
    y = np.random.rand()  # Se genera número aleatorio con $g$
    #print("número generado: ", y)
    criterio = 3.75*y*(1-y)
    #print("f/(cg) = ", criterio)
    u = np.random.rand() # se genera número aleatorio para u
    #print("uniforme = ", u)
    if criterio>u:
      #print('Se acepta:', y)
      k += 1
      X.append(y)
  print('Total de iteraciones = ',j)
  return X
```

```python
Z1 = np.random.randn(10000)
Z2 = np.random.randn(10000)
Z3 = np.random.randn(10000)
V = Z1**2+Z2**2+Z3**2
plt.hist(V, bins=20)
stats.kstest(V, 'chi2', args=(3,0,1))
```

Si $Z \sim N(0,1)$ y $V\sim \chi^2(n)$ entonces 

$$T = \frac{Z}{\sqrt{V/n}} \sim t_{(n)}$$

```python
Z = np.random.randn(1000)
V = np.random.chisquare(10, 1000)
T = Z/np.sqrt(V/10)
plt.hist(T, bins=20)
stats.kstest(T, 't', args=(10,0,1))
```
